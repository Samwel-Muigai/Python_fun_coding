```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

df = pd.read_csv(r'C:\Users\muiga\OneDrive\Desktop\School work\Datadriven.csv')

df.isnull().sum()
```

```python
df = df.ffill()
```

```python
corr_matrix = df.corr(numeric_only=True)
print("/nCorrelation Matrix")
print(corr_matrix)
```

```python
sns.heatmap(corr_matrix, 
            annot=True,
            cmap='coolwarm')
plt.show()
```
